```python
import pandas as pd
import pickle

fname = "C:\\Repositories\\python\\bpfe\\data\\stats\\stats.pkl"
with open(fname, 'rb') as ifile:
    stats = pickle.load(ifile)
```

```python
data = []
for stat in stats:
    row = []
    row.append(stat.finetuning[0].best_test_loss)
    row.append(stat.finetuning[0].best_validation_loss)
    row.append(stat.finetuning.learning_rate)
    row.append(stat.hidden_layers[0].learning_rate)
    row.append(len(stat.hidden_layers))
    row.append(stat.hidden_layers[0].num_nodes)
    data.append(row)
    
df = pd.DataFrame(data, columns=['test_loss', 'validation_loss', 'ft_learning_rate', 'pt_learning_rate', 'num_layers', 'num_nodes'])
```

```python
df['total_loss'] = (df['test_loss'] + df['validation_loss'])/2
df = df.sort(['total_loss'], ascending=[True])
df.head(n=100)
```
